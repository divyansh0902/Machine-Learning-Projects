```python
from sklearn.cluster import KMeans
```

```python
#removing labels column
x = data.drop(['label'], axis=1) #for unsupervised learning no need for labels

#selecting all values of data
x = x.values

print(x.shape)
```

```python
#now determine the optimum number of clusters within dataset by using elbow method

plt.rcParams['figure.figsize'] = (10,4)

wcss = []
for i in range(1,11):
    km = KMeans(n_clusters= i, init= 'k-means++', max_iter=300, n_init=10, random_state=0)
    km.fit(x)
    wcss.append(km.inertia_)
    
#plot the result
plt.plot(range(1,11), wcss)
plt.title('Ther Elbow Method', fontsize= 20)
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
#implement the k means algorithm to perform clustering analysis
#number of cluster = 4
km = KMeans(n_clusters= 4, init= 'k-means++', max_iter=300, n_init=10, random_state=0)
y_means = km.fit_predict(x)

#find the results
a = data['label']
y_means = pd.DataFrame(y_means)
z = pd.concat([y_means, a], axis=1)
z = z.rename(columns = {0: 'cluster'})

#check cluster for each crops
print("lets check the results after applying the K Means clustering anaylsis \n")
print('crops in first cluster:', z[z['cluster'] == 0]['label'].unique())
print('crops in second cluster:', z[z['cluster'] == 1]['label'].unique())
print('crops in third cluster:', z[z['cluster'] == 2]['label'].unique())
print('crops in fourth cluster:', z[z['cluster'] == 3]['label'].unique())
```

```python
#lets split the dataset for predictive modelling

y = data['label']
x = data.drop(['label'], axis=1)

print('shape of x:', x.shape)
print('shape of y:', y.shape)
```

```python
#create training and testing sets for validation of results
from sklearn.model_selection import train_test_split

x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state= 0)

print("shape of x train:", x_train.shape)
print("shape of x test:", x_test.shape)
print("shape of y train:", y_train.shape)
print("shape of y test:", y_test.shape)
```

```python
#create a predictive model

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
#evaluate model performance
from sklearn.metrics import confusion_matrix

#print confusion matrix
plt.rcParams['figure.figsize'] = (10,10)
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot= True, cmap = 'Wistia')
plt.title('confusion matrix for logistic regression', fontsize= 15)
plt.show()
```

```python
#print classification report
from sklearn.metrics import classification_report
cr = classification_report(y_test, y_pred)
print(cr)
```
