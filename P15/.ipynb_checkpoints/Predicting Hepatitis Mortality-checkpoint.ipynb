```python
#Interquartile Range
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
IQR = q3 -q1 
IQR
```

```python
# get actual datapoint that is an outlier
(df < (q1 -1.5 *IQR ))|(df < (q3 +1.5 *IQR ))
```

```python
df_no_outlier = df[~((df < (q1 -1.5 *IQR ))|(df > (q3 +1.5 *IQR ))).any(axis=1)]
```

# feature selection

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
df.columns
```

```python
xfeatures = df[[ 'age', 'sex', 'steroid', 'antivirals', 'fatigue', 'malaise',
       'anorexia', 'liver_big', 'liver_firm', 'spleen_palpable', 'spiders',
       'ascites', 'varices', 'bilirubin', 'alk_phosphate', 'sgot', 'albumin',
       'protime', 'histology']]
ylabels = df['class']
```

```python
skb = SelectKBest(score_func=chi2,k=10)
best_feature_fit = skb.fit(xfeatures,ylabels)
```

```python
print("Scores", best_feature_fit.scores_) #highest score shows best fit
```

```python
bf_02 = best_feature_fit.transform(xfeatures)
bf_02
```

```python
#mapping to feature name
feature_scores = pd.DataFrame(best_feature_fit.scores_,columns=["Feature_Scores"])
```

```python
feature_scores
```

```python
feature_column_names = pd.DataFrame(xfeatures.columns,columns=['Feature_name'])
best_fit_df = pd.concat([feature_scores,feature_column_names],axis=1)
```

```python
best_fit_df
```

```python
best_fit_df.nlargest(12,'Feature_Scores')
```

```python
#Recursive feature elimination
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
logit =  LogisticRegression()
```

```python
rfe = RFE(logit,step=8)
rfe_fir=rfe.fit(xfeatures,ylabels)
```

```python
selected_features = pd.DataFrame(rfe_fir.support_,columns=['Selected_features'])
ranking_features = pd.DataFrame(rfe_fir.ranking_,columns=['ranking_features'])
```

```python
rfe_feature_df = pd.concat([feature_column_names,selected_features,ranking_features],axis=1)
```

```python
rfe_feature_df
```

####
- RFE will eliminate the lowest ranking or true

```python
## checking for feature importance
from sklearn.ensemble import ExtraTreesClassifier
```

```python
et_clf = ExtraTreesClassifier()
et_clf.fit(xfeatures,ylabels)
```

```python
print(et_clf.feature_importances_)
```

```python
feature_importance_df = pd.Series(et_clf.feature_importances_,index=xfeatures.columns)
```

```python
feature_importance_df
```

```python
feature_importance_df.nlargest(12).plot(kind='bar');
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
xfeatures_best = df[[ 'age', 'sex', 'steroid', 'antivirals', 'fatigue','spiders',
       'ascites', 'varices', 'bilirubin', 'alk_phosphate', 'sgot', 'albumin',
       'protime', 'histology']]
```

```python
x_train,x_test,y_train,y_test =train_test_split(xfeatures,ylabels,test_size=0.33,random_state=42)
```

```python
#train/test dataset for best features
x_train_b,x_test_b,y_train_b,y_test_b =train_test_split(xfeatures_best,ylabels,test_size=0.33,random_state=42)
```

```python
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
# model accuracy
# method 1
logreg.score(x_test,y_test)
```

```python
#method 2
logreg.predict(x_test)
```

```python
#model accuracy
#method 1
model_logit.score(x_test_b,y_test_b)
```
